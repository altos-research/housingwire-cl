```python
import pandas as pd
from numpy import inf, nan
```

```python
overview = pd.read_csv('./raw/hw_dallas_geos_overview.csv')
overview.rename(columns={
    'listings':'Listings',
    'avg_ppsf':'Avg. PPSF',
    'avg_dom':'Avg. Days on Market',
    'avg_sales_to_list':'Avg. List Price:Sales Price Ratio',
    'offices':'Offices'
}, inplace=True)

pivoted = overview.pivot(index='listing_year', columns='geo')

df = pd.DataFrame()
for col_name in overview.columns[2:]:
    grouped = pivoted.copy()[col_name]
    grouped['group'] = col_name 
    df = pd.concat([df, grouped])

for col_name in ['Listings', 'Avg. PPSF', 'Avg. Days on Market', 'Avg. Sales Price:List Price Ratio', 'Offices']:
    try:
        df.loc[df.group==col_name, ['Dallas', 'Rest of TX', 'Rest of US']] = \
        df.loc[df.group==col_name, ['Dallas', 'Rest of TX', 'Rest of US']].div(df.loc[(df.group==col_name)&(df.index==2019), ['Dallas', 'Rest of TX', 'Rest of US']].iloc[0])
    except:
        df.loc[df.group==col_name, ['Dallas', 'Rest of TX', 'Rest of US']] = \
        df.loc[df.group==col_name, ['Dallas', 'Rest of TX', 'Rest of US']].div(df.loc[(df.group==col_name)&(df.index==2019), ['Dallas', 'Rest of TX', 'Rest of US']])
df.to_csv('./cleaned/GeosOverview.csv')
```

```python
dal = pd.read_csv('./raw/housingwire_datadigest_dallas.csv')
dal.head()
```

```python
dal.sort_values(by='total_listings', ascending=False).head(50)
```

```python
common_brokerage_names = [
    ### note the spaces after REDFIN and EXP
    'COLDWELL BANKER', 'CENTURY 21', 'REDFIN ', 
    'KELLER WILLIAMS', 'EXP ', 'HOMESMART', 
    'COMPASS', 'RE/MAX', 'BERKSHIRE HATHAWAY', 
    'OPENDOOR', 'OFFERPAD', 'ZILLOW', 'BETTER HOMES & GARDENS',
    'DOUGLAS ELLIMAN', 'PPMG'
]

dallas_common_names = [
    'DAVE PERRY MILLER', 'EBBY HALLIDAY', 'ALLIE BETH ALLMAN', 'ROGERS HEALY', 'JPAR', 'WILLIAM DAVIS',
    'CLAY STAPP', 'DAVID GRIFFIN', 'DAVID M WEEKLEY', 'JOSH DESHONG', 'MEYER GROUP', 'MIKE MAZYCK',
    'ULTRA REAL'
]

common_brokerage_names += dallas_common_names

dal.listing_office_name.fillna('', inplace=True)
dal['listing_office_name'] = dal.listing_office_name.str.upper()\
                                    .str.rstrip()\
                                    .str.replace('.','')\
                                    .str.replace(',','')\
                                    .str.replace("'",'')\
                                    .str.replace('LLC', '')\
                                    .str.replace('INC', '')
dal['listing_office_name'] = dal.listing_office_name.str.strip()
### SPECIAL: clean SOTHEBY'S INTERNATIONAL REALTY
dal['listing_office_name'] = [x if 'SOTHEBY' not in x 
                                    else 'SOTHEBY\'S INTERNATIONAL REALTY'
                                    for x in dal.listing_office_name]
### SPECIAL: clean RE/MAX
dal['listing_office_name'] = [x if 'REMAX' not in x 
                                    else 'RE/MAX' 
                                    for x in dal.listing_office_name]
### SPECIAL: clean C21 
dal['listing_office_name'] = [x if 'C21' not in x 
                                    else 'CENTURY 21' 
                                    for x in dal.listing_office_name]
### SPECIAL: JP & Associates 
dal['listing_office_name'] = [x if 'JP & ASSOCIATES' not in x 
                                    else 'JPAR' 
                                    for x in dal.listing_office_name]

new_names = []
for co in dal['listing_office_name']:
    added = False
    for name in common_brokerage_names:
        if co.startswith(name):
            new_names.append(name)
            added = True
    if not added:
        new_names.append(co)

dal['listing_office_group'] = new_names
```

```python
byCo = dal.groupby(['listing_year','listing_office_group']).agg({
    'listing_office_name':'count',
    'total_listings':'sum',
    'avg_price_per_square_foot':'mean',
    'avg_derived_days_on_market':'mean',
    'avg_sales_price_to_list_price_ratio':'mean'
}).reset_index().rename(columns={
    'listing_office_name':'offices'
})
byCo
```

```python
leaders = byCo.loc[byCo.total_listings>99]
leaders
```

```python
leadersListings = leaders[['listing_year', 'listing_office_group', 'total_listings']].pivot(columns='listing_year', index='listing_office_group', values='total_listings')
leadersListings['one_yr'] = round((leadersListings[2023] / leadersListings[2022] - 1) * 100, 3)
leadersListings['five_yr'] = round((leadersListings[2023] / leadersListings[2019] - 1) * 100, 3)
leadersListings['2019_share'] = round(leadersListings[2019] / leadersListings[2019].sum() * 100, 3)
leadersListings['2020_share'] = round(leadersListings[2020] / leadersListings[2020].sum() * 100, 3)
leadersListings['2021_share'] = round(leadersListings[2021] / leadersListings[2021].sum() * 100, 3)
leadersListings['2022_share'] = round(leadersListings[2022] / leadersListings[2022].sum() * 100, 3)
leadersListings['2023_share'] = round(leadersListings[2023] / leadersListings[2023].sum() * 100, 3)
leadersListings['one_yr_share'] = round((leadersListings['2023_share'] / leadersListings['2022_share'] - 1) * 100, 3)
leadersListings['five_yr_share'] = round((leadersListings['2023_share'] / leadersListings['2019_share'] - 1) * 100, 3)
leadersListings.sort_values(by=['2023_share','2022_share','2021_share','2020_share','2019_share'], ascending=False).head(20)
```
